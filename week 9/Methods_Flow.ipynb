## Anchors Flow Explanation (Step-to-Step) (Green Chart)

* black-box(step 1): started with a trained, complex ML model, such as the XGBoost.

* Instance to be explanied(step 2): Choose a single instance whose prediction you want to understand.

* Generate local neighborhood samples(step 3): Generated perturbed samples around the selected instance on step 2 (instance).

* Get prediction from complex models on samples(step 4): Fed the generated sample into the beginning of the flow which is the original complex model.

* Identify Anchor Rules(step 5): In this step, we can identify simple and interpretable IF-THEN rules from the perturbations we obtain on step 3.

* FPrecision Check(step 6): Verify the anchor rules against a precision threshold to ensure their reliability.

* Explanation(step 7): Provide a detailed explanation on the anchor rules for the prediction.

## SHAP Flow Explanation (Step-to-Step) (Grey Chart)

* black-box(step 1): started with a trained, complex ML model, such as the XGBoost.

* Instance to be explanied(step 2): Choose a single instance whose prediction you want to understand.

* Generate local neighborhood samples(step 3): Generated perturbed samples around the selected instance on step 2 (instance).

* Get prediction from complex models on samples(step 4): Fed the generated sample into the beginning of the flow which is the original complex model.

* Compute the SHAP Values(step 5): This step is very significant in the SHAP method since it will calculate Shapley values using cooperative game theory. And this can fairly distribute contributions to each feature, attributing their impact on the prediction.

* Feature Contriution Identification(step 6): Identify and quantify the contribution of each feature based on the computed Shapley values.

* Explanation(step 7): Provide a detailed explanation of the model's prediction for the chosen. 

## LIME Flow Explanation (Step-to-Step) (Orange Chart)

* black-box(step 1): refers to the ML models that's accurate but not easy to interpretable. Such as deep neural networks, XFBoost

* Instance to be explanied(step 2): Select a specific data point from our dataset which prediction you want to interpret. 

* Generate local neighborhood samples(step 3): Which refers to a slight variations of the chosen instance (usually a small adjustments to the feature values)

* Get prediction from complex models on samples(step 4): This step allow us to fed the generated sample into the beginning of the flow which is the original black-box model.

* fit a simple interpretable local model(step 5): This step is use a simple model like linear regression to approximate the predictions made by the complex model for the neighborhood samples.

* Identify important features(step 6): Use the local interpretable model to determine which features significatly influence the prediction.

* Explanation(step 7): Explain the final output in an understandable way. Very important that is to show the important features that contribute to the prediction of the original model.
